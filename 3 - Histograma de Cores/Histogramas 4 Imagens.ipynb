```python
%matplotlib inline
import cv2 # OpenCV
import matplotlib.pyplot as plt # Matplotlib
import numpy as np # Numpy
```

```python
figure, plots = plt.subplots(ncols=4, nrows=1, figsize = (20, 10))

for x, subplot in zip(["../Imgs/pinguim.jpg", "../Imgs/lenna.jpg", "../Imgs/dark.jpg", "../Imgs/sunny.jpg"], plots):
    aux = cv2.imread(x, cv2.IMREAD_UNCHANGED)
    aux = cv2.cvtColor(aux, cv2.COLOR_BGR2RGB)
    subplot.imshow(aux)
```

```python
for x in ["../Imgs/pinguim.jpg", "../Imgs/lenna.jpg", "../Imgs/dark.jpg", "../Imgs/sunny.jpg"]:
    img = cv2.imread(x)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_cinza = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    img_gs = cv2.cvtColor(img_cinza, cv2.COLOR_GRAY2RGB)
    img_vetor = img_cinza.ravel()
    h = cv2.calcHist([img_vetor], [0], None, [255], [0,255])
    plt.plot(h)
    plt.ylim([0, 10000])
plt.legend(["../pinguim.jpg", "../lenna.jpg", "../dark.jpg", "../sunny.jpg"], loc = 2)
plt.show()
```
