```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread("../Imgs/Pinguins_Madagascar.jpg", 1)
img2 = cv2.imread("../Imgs/nemo.jpeg", 1)
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img1 = cv2.resize(img1, (400, 400))
img2 = cv2.resize(img2, (400, 400))
```

```python
plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
```

```python
img3 = np.zeros(img1.shape, dtype = np.uint8)
```

```python
for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
        for k in range(3):
            img3[i, j, k] += img1[i, j, k] & 0xF0
```

```python
for i in range(img2.shape[0]):
    for j in range(img2.shape[1]):
        for k in range(3):
            img3[i, j, k] += (img2[i, j, k] & 0xF0) >> 4
```

```python
fig, axis = plt.subplots(1, 2, figsize = (20, 20))
[ax.set_axis_off() for ax in axis]

axis[0].imshow(img1)
axis[1].imshow(img3)
```

```python
img4 = np.zeros(img1.shape, dtype = np.uint8)
```

```python
print(img1.dtype)
print(img2.dtype)
print(img3.dtype)
print(img4.dtype)
```

```python
for i in range(img4.shape[0]):
    for j in range(img4.shape[1]):
        for k in range(3):
            img4[i, j, k] = (img3[i, j, k] & 0x0F) << 4
```

```python
plt.imshow(img4)
```

```python
img5 = np.zeros(img1.shape, dtype = np.uint8)
```

```python
for i in range(img4.shape[0]):
    for j in range(img4.shape[1]):
        for k in range(3):
            img5[i, j, k] = (img3[i, j, k] & 0xF0)
```

```python
plt.imshow(img5)
```

```python
fig, axis = plt.subplots(1, 3, figsize = (20, 20))
[ax.set_axis_off() for ax in axis]

# img3 -> img com estenografia (junção de img1 e img2). img4, img5 -> imagens recuperadas de img3
axis[0].imshow(img3)
axis[1].imshow(img4)
axis[2].imshow(img5)
```
