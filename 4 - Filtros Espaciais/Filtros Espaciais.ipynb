```python
%matplotlib inline
import cv2 # OpenCV
import matplotlib.pyplot as plt # Matplotlib
import numpy as np # Numpy
from scipy import ndimage
```

```python
img = cv2.imread("../Imgs/pinguim.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kernel = np.array([[1, 0, -1],[0, 0, 0],[-1, 0, 1]])
img_edges = cv2.filter2D(img, -1, kernel)
plt.imshow(img_edges)
```

```python
kernel = np.array([[0, -1, 0],[-1, 5, -1],[0, -1, 0]])
img_sharpen = cv2.filter2D(img, -1, kernel)
plt.imshow(img_sharpen)
```

```python
from skimage.util import random_noise

img_sp = random_noise(img, mode = "s&p")
plt.imshow(img_sp)
```

```python
img_mediana = cv2.medianBlur(np.float32(img_sp), 5)
plt.imshow(img_mediana)
```

```python
import numpy as np
import cv2
import importar
from importlib import reload

cap = cv2.VideoCapture(0)



print(cap.isOpened())


while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret==True:
        reload(importar)
        
        try:
            cv2.imshow('Frame', importar.salt_pepper_half(frame))
        except Exception as e:
            importar.default(frame, e)
        

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Release everything if job is finished
cap.release()
cv2.destroyAllWindows()
```
