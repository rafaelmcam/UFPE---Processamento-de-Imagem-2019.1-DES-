```python
import numpy as np
import cv2

img = cv2.imread('../Imgs/pinguim.jpg')

##img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Z = img.reshape((-1,3))

# convert to np.float32
Z = np.float32(Z)

# define criteria, number of clusters(K) and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 4
ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)

# Now convert back into uint8, and make original image
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((img.shape))

cv2.imshow('res2', res2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from matplotlib import cm
from matplotlib import colors

nemo = cv2.imread('../Imgs/nemo.jpeg')
nemo = cv2.cvtColor(nemo, cv2.COLOR_BGR2RGB)
plt.imshow(nemo)


cinza = (0, 0, 150)
branco = (255, 30, 255)


cinza_img = np.full((10, 10, 3), cinza, dtype=np.uint8) / 255.0
cinza_hsv = colors.rgb_to_hsv(cinza_img)

branco_img = np.full((10, 10, 3), branco, dtype=np.uint8) / 255.0
branco_hsv = colors.rgb_to_hsv(branco_img)

hsv_nemo = cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)

cinza_hsv = (0, 0, 200)
branco_hsv = (145, 60, 255)

# plt.subplot(1, 2, 1)
# plt.imshow(colors.hsv_to_rgb(cinza_hsv))
# plt.subplot(1, 2, 2)
# plt.imshow(colors.hsv_to_rgb(branco_hsv))
# plt.show()
```

```python
mask = cv2.inRange(hsv_nemo, cinza_hsv, branco_hsv)
result = cv2.bitwise_and(nemo, nemo, mask=mask)

plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
